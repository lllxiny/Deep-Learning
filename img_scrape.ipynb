```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import urllib.request
import time

def save_google_images(query, num_images):
    # Initialize Chrome WebDriver (replace 'path_to_chromedriver' with the actual path)
    driver = webdriver.Chrome(executable_path='/Users/liuxinyu/chromedriver_mac_arm64/chromedriver')

    try:
        # Go to Google Images
        driver.get('https://www.google.com/imghp')

        # Enter the search query
        # search_box = driver.find_element(By.CSS_SELECTOR, 'input[name="q"]')
        # search_box.send_keys(query)
        # search_box.submit()

        search_box = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea')
        search_box.send_keys(query)
        search_box.submit()

        # Scroll to load more images
        scroll_pause_time = 2
        last_height = driver.execute_script('return document.body.scrollHeight')
        while len(driver.find_elements(By.CSS_SELECTOR, 'img.Q4LuWd')) < num_images:
            driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
            time.sleep(scroll_pause_time)
            new_height = driver.execute_script('return document.body.scrollHeight')
            if new_height == last_height:
                break
            last_height = new_height

        # Extract and save image URLs
        image_urls = driver.find_elements(By.CSS_SELECTOR, 'img.Q4LuWd')
        for i, image in enumerate(image_urls[:num_images]):
            image_url = image.get_attribute('src')
            image_file_name = f'zzuucchhhhiinni_{i+1}.jpg'

            if image_url:
                urllib.request.urlretrieve(image_url, image_file_name)
                print(f'Saved image: {image_file_name}')
            else:
                print(f'Skipped invalid image URL for index: {i+1}')

    finally:
        # Quit the WebDriver
        driver.quit()

# Call the function to save 15e images of the "onion image" query
save_google_images('raw whole zucchini one white background', 50000)
```

```python
import os

def rename_images(folder_path, vegetable_name):
    # Get a list of all files in the folder
    files = os.listdir(folder_path)

    # Iterate through each file
    for i, file_name in enumerate(files, start=1):
        # Split the file name and extension
        name, extension = os.path.splitext(file_name)

        # Construct the new file name
        new_file_name = f"{vegetable_name}_{i}{extension}"

        # Get the full path of the file
        old_file_path = os.path.join(folder_path, file_name)
        new_file_path = os.path.join(folder_path, new_file_name)

        # Rename the file
        os.rename(old_file_path, new_file_path)

        print(f"Renamed {file_name} to {new_file_name}")

# Specify the folder path and vegetable name
folder_path = "/Users/liuxinyu/final proj/garlic"
vegetable_name = "garlic"

# Call the function to rename the images
rename_images(folder_path, vegetable_name)
```

```python
import os
import random
import shutil

def randomly_assign_images(folder_path, num_images, destination_folder):
    # Get a list of all image files in the folder
    image_files = [file for file in os.listdir(folder_path) if file.endswith(('.jpg'))]

    # Shuffle the image files randomly
    random.shuffle(image_files)

    # Select the first num_images from the shuffled list
    selected_images = image_files[:num_images]

    # Create the destination folder if it doesn't exist
    os.makedirs(destination_folder, exist_ok=True)

    # Move the selected images to the destination folder
    for image_file in selected_images:
        source_path = os.path.join(folder_path, image_file)
        destination_path = os.path.join(destination_folder, image_file)
        shutil.move(source_path, destination_path)

    print(f"{num_images} images have been randomly assigned to the '{destination_folder}' folder.")

# Usage example
randomly_assign_images('/Users/liuxinyu/final proj/img/zucchini_a', 150, '/Users/liuxinyu/final proj/img/zucchini_preprocessed_validation')
```

```python
import os
import random
import shutil

def randomly_assign_images(folder_path, num_images, destination_folder):
    # Get a list of all image files in the folder
    image_files = [file for file in os.listdir(folder_path) if file.endswith(('.jpg', '.png', '.jpeg'))]

    # Shuffle the image files randomly
    random.shuffle(image_files)

    # Select the first num_images from the shuffled list
    selected_images = image_files[:num_images]

    # Create the destination folder if it doesn't exist
    os.makedirs(destination_folder, exist_ok=True)

    # Move the selected images to the destination folder and remove them from the original folder
    for image_file in selected_images:
        source_path = os.path.join(folder_path, image_file)
        destination_path = os.path.join(destination_folder, image_file)
        shutil.move(source_path, destination_path)

    print(f"{num_images} images have been randomly assigned to the '{destination_folder}' folder.")

# Usage example
randomly_assign_images('/Users/liuxinyu/final proj/img/zucchini_a', 600, '/Users/liuxinyu/final proj/img/zucchini_preprocessed_train')

```
