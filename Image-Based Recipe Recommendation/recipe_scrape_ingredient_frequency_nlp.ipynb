# Chinese recipes

```python
#chinese recipes
chn_url = 'https://www.allrecipes.com/recipes/695/world-cuisine/asian/chinese/'
chn_recipe_all = requests.get(chn_url)
chn_recipe_results = chn_recipe_all.text

with open(f'./chn_recipe.html','w', encoding='utf8') as fullPage:
    fullPage.write(chn_recipe_results)
    print('Recipe result page printed')
```

```python
chn_soup = soup(open(f'./chn_recipe.html','r').read(), 'lxml')
```

```python
chn_recipe_div = chn_soup.find('div',{'class':'comp mntl-taxonomysc-article-list-group mntl-block'})
```

```python
print(chn_recipe_div)
```

```python
chn_recipe_all_1 = soup(chn_recipe_all.text,'html.parser')
chn_recipe_all_1
```

```python
chn_recipe_try1 = chn_recipe_all_1.select('div.comp.tax-sc__recirc-list.card-list.mntl-document-card-list.mntl-card-list.mntl-block > a')
chn_recipe_try1
```

```python
chn_recipe_urls = re.findall('href="https:\/\/www.allrecipes.com\/recipe\/.+\/', str(chn_recipe_try1))
chn_recipe_urls
```

```python
chn_recp_url = re.sub('href="','', str(chn_recipe_urls))
chn_recp_url = re.sub('\'','', str(chn_recp_url))
chn_recp_url = re.sub('\,','', str(chn_recp_url))
chn_recp_url = re.sub('\[','', str(chn_recp_url))
chn_recp_url = re.sub('\]','', str(chn_recp_url))
chn_recp_url_list = chn_recp_url.split()
chn_recp_url_list
```

```python
recipe_name_pattern = r'href="https://www.allrecipes.com/recipe/\d+/(.*)/'
chn_recipe_names = []

for i in chn_recipe_urls:
    match = re.search(recipe_name_pattern, i)
    if match:
        recipe_name = match.group(1)
        print(recipe_name)
        chn_recipe_names.append(recipe_name)
```

```python
chn_recipe_names_clean = []

for text in chn_recipe_names:
    modified_text = re.sub(r'-', ' ', str(text))
    chn_recipe_names_clean.append(modified_text)



print(chn_recipe_names_clean)
```

```python
chn_ingre_li = []

for i in range(len(chn_recp_url_list)):
    url1 = chn_recp_url_list[i]
    url1_all = requests.get(url1)
    url1_all_results = url1_all.text

    with open(f'./chn_recipe + {i+1}.html','w', encoding='utf8') as fullPage:
        fullPage.write(url1_all_results)
        print('Recipe result page printed')

    url1_soup = soup(open(f'./chn_recipe + {i+1}.html','r').read(), 'lxml')

    chn_ingre_all_1 = soup(url1_all.text,'html.parser')

    # ingredient_name_pattern = r'<span data-ingredient-name="true">(.+)'

    chn_ingre = re.findall('<span data-ingredient-name="true">(.+)', str(chn_ingre_all_1))
    chn_ingre = re.sub('</span></p>','',str(chn_ingre))

    chn_ingre_li.append(chn_ingre)

print(chn_ingre_li)
```

```python
chn_nation = []
for i in range(len(chn_recipe_names_clean)):
    chn_nation.append('Chinese')
```

```python
chn_df = pd.DataFrame({'name': chn_recipe_names_clean, 'ingredient': chn_ingre_li, 'region': chn_nation, 'link': chn_recp_url_list})

print(chn_df)
```
